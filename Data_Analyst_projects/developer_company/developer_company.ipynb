###  Подсчет для каждой кампании количества обращений в каждый из дней

```python
# Создадим необходимую сводную таблицу по аналогии
campaign_2 = data[data['статус_обращения'] == 'Успешно'].pivot_table(index = 'utm_campaign', columns = 'дата_работы', 
                                                                     values = 'статус_обращения', aggfunc = 'count')
campaign_2['total'] = campaign_2['2021-03-01'] + campaign_2['2021-03-02'] + campaign_2['2021-03-02']
campaign_2.sort_values(by = 'total', ascending = False)
```

```python
# Визуализируем полученные данные
data[data['статус_обращения'] == 'Успешно'].pivot_table(index = 'дата_работы', columns = 'utm_campaign', 
                                                        values = 'статус_обращения', aggfunc = 'count')\
                .plot(kind = 'bar', title = 'Количество успешных обращений', figsize = (8, 5), rot = 0)
plt.legend(fontsize=8)
plt.show()
```

### Определение успешности кампаний

Однако для полноты анализа необходимо посмотреть конверсию успешных посещений, а также учесть стоимость рекламных кампаний.

```python
# Создадим новую таблицу из предыдущих
table = pd.DataFrame({'всего_обращений' : campaign['total'],
                      'успешных_обращений' : campaign_2['total']})
# Добавим суммарную стоимость рекламной кампании
table = table.merge(data_2.pivot_table(index = 'utm_campaign', values = 'расход_руб', aggfunc = 'sum'), on = 'utm_campaign')
table.rename(columns = {'расход_руб' : 'стоимость'}, inplace = True)

# Оценим конверсию и выполним сортировку
table['CR'] = round(table['успешных_обращений'] / table['всего_обращений'], 2)
table.sort_values(by = 'CR', ascending = False)
```

```python
# Посчитаем условную стоимость успешного обращения
table['SC'] = round(table['стоимость'] / table['успешных_обращений'], 2)
table.sort_values(by = 'SC')
```

<div style="border:solid Green 2px; padding: 20px">

**Вывод:**

Таким образом получились следующие результаты:
- Самая большая конверсия отмечается у кампаний номер 1, 6 и 8. Причем стоит отметить, что явным лидером является кампания 1 с показателем 0.68.
- А самыми выгодными кампаниями с учетом расходов являются кампании 8, 1 и 4. В данном случае лидерство с большим отрывом забирает кампания 8.

Учитывая удельные и абсолютные показатели можно **в значительной степени выделить рекламную кампанию номер 8** - она имеет наименьшую стоимость при других сравнимых показателях.

## Задание 3: 

### Распределение обращений в течение дня

```python
# Подготовим таблицу - создадим дополнительный столбец с временем обращения в строковом типе данных
data.insert(loc=5, column = 'время_обращения', 
              value = pd.to_datetime(data['дата_создания']).dt.time.astype('str'))
```

```python
# Создадим функцию для категоризации времени обращения
def times (time):
    if '21:00:00' <= time <= '23:59:59':
        return 'ночь'
    if '00:00:00' <= time < '06:00:00':
        return 'ночь'
    if '06:00:00' <= time < '10:00:00':
        return 'утро'
    if '10:00:00' <= time < '17:00:00':
        return 'день'
    if '17:00:00' <= time < '21:00:00':
        return 'вечер'
```

```python
# Применим функцию к столбцу со временем обращения
data.insert(loc=6, column = 'время_суток', 
              value = data['время_обращения'].apply(times))
```

```python
data.head()
```

```python
# Рассмотрим количество обращений в течение суток
time_table = data.groupby(by = 'время_суток')['utm_campaign'].agg('count')
time_table.sort_values(ascending = False)
```

```python
time_table.plot(kind = 'bar', title = 'Обращения в течение суток', ylabel = 'Количество обращений')
plt.xticks (rotation = 0)
plt.show()
```

<div style="border:solid Green 2px; padding: 20px">

**Вывод:**

Больше всего обращений поступает в течение дня, но что удивительно, на втором месте находится ночное время суток - довольно много клиентов обращаются ночью. Меньше всего обращений поступает утром - с 6 до 10 утра.

### Распределение обращений в течение суток с учетом разделения рекламных кампаний

```python
# Отобразим в виде гистограммы распределение обращений в течение суток по кампаниям
data.pivot_table(index = 'время_суток', columns = 'utm_campaign', values = 'объект', aggfunc = 'count')\
    .plot(kind = 'bar', figsize = (8, 6))

plt.legend(loc = 'upper right', fontsize=9)
plt.xticks (rotation = 0 )
plt.show()
```

<div style="border:solid Green 2px; padding: 20px">

**Вывод:**

Тенденция увеличения количества обращений днем и снижение утром сохраняется в разрезе всех кампаний. Отличительной особенностью характеризуется кампания номер 2 - в разрезе нее наблюдается наибольшее количество обращений в ночное время суток.
